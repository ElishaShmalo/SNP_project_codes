Intall and call the package

```python
from snp.eda import EDARunner, MultiEDARunner
import pandas as pd
import matplotlib.pyplot as plt
```

Read the determinist dataset that has no noise.

```python
df_deterministic = pd.read_csv('synthetic_data/Deterministic.csv')
df_deterministic['Timestamp'] = pd.to_datetime(df_deterministic['Timestamp'])
EDA_deterministic = EDARunner(df_deterministic,"Deterministic | Mean=0 | Std=0")
```

```python
EDA_deterministic.overview(Zoom_in=(0,1000))
```

```python
EDA_deterministic.plot_decomposition()
```

Read the dataset that holds frequency and phase constant, adding noise to amplitude.

Add different noise to the sine wave.

```python
df_amp_noise = pd.read_csv('synthetic_data/AmpNoise-Mean0-Slope0p1-StdAllSDValues.csv')
df_amp_noise['Timestamp'] = pd.to_datetime(df_amp_noise['Timestamp'])
EDA_amp_noise = MultiEDARunner(df_amp_noise, "Amplitude Noise",colormap=plt.cm.Blues)
```

```python
df_freq_noise = pd.read_csv('synthetic_data/FreqNoise-Mean0-Slope0p1-StdAllSDValues.csv')
df_freq_noise['Timestamp'] = pd.to_datetime(df_freq_noise['Timestamp'])
EDA_freq_noise = MultiEDARunner(df_freq_noise, "Frequency Noise",colormap=plt.cm.Greens)
```

```python
df_phase_noise = pd.read_csv('synthetic_data/PhaseNoise-Mean0-Slope0p1-StdAllSDValues.csv')
df_phase_noise['Timestamp'] = pd.to_datetime(df_phase_noise['Timestamp'])
EDA_phase_noise = MultiEDARunner(df_phase_noise, "Phase Noise",colormap=plt.cm.Reds)
```

### Amplitude Noise
An overview of the whole data and a zoomed in view. The sine wave becomes less recognizable as the noise increase.

```python
EDA_amp_noise.overview(columns=df_amp_noise.columns[[1, 2, 10, 20]])
EDA_amp_noise.overview(Zoom_in=(0,1000),columns=df_amp_noise.columns[[1, 2, 10, 20]])
```

### Frequency Noise
An overview of the whole data and a zoomed in view. The noise seem to accumulate as time increase.

```python
EDA_freq_noise.overview(columns=df_freq_noise.columns[[1, 2, 10, 20]])
EDA_freq_noise.overview(Zoom_in=(0,1000),columns=df_freq_noise.columns[[1, 2, 10, 20]])
```

### Amplitude Noise
An overview of the whole data and a zoomed in view. The sine wave becomes less recognizable as the noise increase.

```python
EDA_phase_noise.overview(columns=df_phase_noise.columns[[1, 2, 10, 20]])
EDA_phase_noise.overview(Zoom_in=(0,1000),columns=df_phase_noise.columns[[1, 2, 10, 20]])
```

### variance and standard deviation
The variance and standard deviation for amplitude noise increase as noise increases. Frequency and phase noise does not change as noise increases.

```python
EDA_amp_noise.visualize_variance_std_dev()
EDA_freq_noise.visualize_variance_std_dev()
EDA_phase_noise.visualize_variance_std_dev()
```

### Autocorrelation accross datasets
